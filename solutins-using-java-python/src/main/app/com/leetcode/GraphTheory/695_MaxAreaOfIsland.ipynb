```python
class Solution:
    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        def dfs(grid, i, j):
            if grid[i][j] == 0:
                return 0
            
            earth = 1
            grid[i][j] = 0
            if i + 1 < len(grid):
                earth += dfs(grid, i + 1, j)
            if i - 1 >= 0:
                earth += dfs(grid, i - 1, j)
            if j + 1 < len(grid[0]):
                earth += dfs(grid, i, j + 1)
            if j - 1 >= 0:
                earth += dfs(grid, i, j - 1)
            
            return earth
        
        m = len(grid)
        n = len(grid[0])
        maxi = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    maxi = max(maxi, dfs(grid, i, j))
        
        return maxi
```
